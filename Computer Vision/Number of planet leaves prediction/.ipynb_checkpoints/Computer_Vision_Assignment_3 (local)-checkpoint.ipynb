# Load Data

```python
def load_images(data_path):
    #store train and test datasets in one array
  images = []
  classes = []
  dirs = os.listdir(data_path)
  # looping over all selected folders using dirs array
  for folder_name in sorted(dirs):
      
      #store each image in its specified folder
      for file in os.listdir(data_path+'/'+folder_name):
          image = io.imread(data_path+'/'+folder_name+'/'+file)
          image = cv2.resize(image, (128,128))
          image = image.astype(np.int)
          #all_data[0].append()
          #folder_data.append([image,int(folder_name)])
          # flatten the images to easily deal with it
          #image = np.reshape(image,-1)
          # store image and its class in a dataframe
          folder = int(folder_name)
          images.append(image)
          classes.append(folder)
          #dictt = {'image':image,'class':folder}
          #df = df.append(dictt,ignore_index=True)        
  return images,classes
```

```python
train_dir = 'training'
test_dir = 'testing'

train_x,train_y = load_images(train_dir)
X_test,y_test = load_images(test_dir)
```

# Transfer Learning

## Import VGG

```python
from tensorflow.keras.applications import VGG16
vgg_model = VGG16(weights='imagenet',
                  include_top=False,
                  input_shape=(128, 128, 3))
```

```python
print(vgg_model.summary())
```

```python
'''model= tf.keras.Sequential()
for layer in vgg_model.layers[0:7]:
      model.add(layer)
for layer in model.layers:
        layer.trainable = False
model.summary()
model.add(Conv2D(64, (3, 3), activation='relu', padding='same', name='block3_conv1') )
model.add( Conv2D(64, (3, 3), activation='relu', padding='same', name='block3_conv2') )
model.add( MaxPooling2D((2, 2), strides=(2, 2), name='block3_pool'))
for layer in vgg_model.layers[7:-1]:
      model.add(layer)'''
```

```python
for layer in vgg_model.layers[0:7]:
        layer.trainable = False

x = vgg_model.layers[-13].output
x = Conv2D(50, 9, activation='relu', padding='same', name='block3_conv1')(x)
x = Conv2D(50, 9, activation='relu', padding='same', name='block3_conv2')(x)
x = MaxPooling2D((8, 8), strides=(2, 2))(x)
#x = Conv2D(512, 4, activation='relu', padding='same', name='block4_conv1')(x)
#x = Conv2D(512, 4, activation='relu', padding='same', name='block4_conv2')(x)
#x = MaxPooling2D((, 5), strides=(2,2))(x)
x = Flatten()(x)
x = Dense(5000,activation='relu')(x)
x = Dense(100,activation='relu')(x)
x = Dense(5, activation='relu')(x)
vgg_model_transfer = Model(inputs=vgg_model.input, outputs=x)
print(vgg_model_transfer.summary())
```

```python
def transform_labels(arr):
  for i in range(len(arr)):
    if arr[i] == 6:
      arr[i]= 4
    elif arr[i] == 4:
      arr[i]= 3
    elif arr[i] == 3:
      arr[i]= 2
    elif arr[i] == 2:
      arr[i]= 1
    else: 
      arr[i]= 0
  return arr
```

```python
y_train = transform_labels(y_train)
y_validate = transform_labels(y_validate)
y_test = transform_labels(y_test)
```

```python
nClasses = 5
# Use Keras' handy utils
y_train_k = tensorflow.keras.utils.to_categorical(y_train, num_classes=nClasses)
y_validate_k = tensorflow.keras.utils.to_categorical(y_validate, num_classes=nClasses)
y_test_k = tensorflow.keras.utils.to_categorical(y_test, num_classes=nClasses)
```

## Training

```python
X_train = np.array(X_train)
X_validate = np.array(X_validate)
X_test = np.array(X_test)
```

```python
batchSize = 10
nEpochs = 15

sgd = tensorflow.keras.optimizers.SGD(lr=0.0001)
vgg_model_transfer.compile(loss='categorical_crossentropy', 
                           optimizer=sgd, 
                           metrics=['accuracy'])
                           
history = vgg_model_transfer.fit(X_train, y_train_k, batch_size=batchSize, epochs=nEpochs, verbose=1, 
                                 validation_data=(X_validate, y_validate_k))
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(nEpochs)

plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
y_predict_prob = vgg_model_transfer.predict(X_test)
y_predict = y_predict_prob.argmax(axis=-1)
```

```python
print(classification_report(y_test, y_predict))
```

```python
from sklearn.metrics import confusion_matrix

conf_mx = confusion_matrix(y_test, y_predict)
print('Confusion matrix: \n{0}'.format(conf_mx))
```

```python
def plot_confusion_matrix(matrix):
    """If you prefer color and a colorbar"""
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111)
    cax = ax.matshow(matrix)
    fig.colorbar(cax)
    
row_sums = conf_mx.sum(axis=1, keepdims=True)
norm_conf_mx = conf_mx / row_sums
np.fill_diagonal(norm_conf_mx, 0)
# plt.matshow(norm_conf_mx, cmap=plt.cm.gray)
plot_confusion_matrix(norm_conf_mx)
plt.show()
```
